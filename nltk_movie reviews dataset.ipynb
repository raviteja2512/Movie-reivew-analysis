```python
from nltk.corpus import movie_reviews
```

```python
movie_reviews.fileids()
```

```python
posr=movie_reviews.fileids('pos')
negr=movie_reviews.fileids('neg')
```

```python
print(movie_reviews.raw(fileids=(posr[0])))
```

```python
k=movie_reviews.words(fileids=(posr[0]))
```

```python
len(k)
```

```python
type(movie_reviews)
```

```python
import string
total_stopwords=nltk.corpus.stopwords.words('english')+list(string.punctuation)
```

```python
total_stopwords
```

```python
def after_removal_of_stopwords(words):
    return {i:1 for i in words 
            if not i in total_stopwords}
```

```python
final_words=after_removal_of_stopwords(k)
```

```python
len(final_words)
```

Now lets do the BAG OF WORDS MODEL with all movie reivews(corpse) and frequency plot

```python
final_MR_words=after_removal_of_stopwords_MR(movie_reviews.words())
```

```python
def after_removal_of_stopwords_MR(words):
    return [i for i in words 
            if not i in total_stopwords]
```

```python
type(final_MR_words)
```

WE CREATED THIS LIST UNLIKE DICTIONARY IN THE PREVIOUS FUNCTION BECAUSE WE ARE FINDING COUNTS USING COUNTER. IT NEEDS WORDS IN THE FORM OF LIST.

```python
len(movie_reviews.words())
```

```python
from collections import Counter

count_of_words=Counter(final_MR_words)
```

```python
type(count_of_words)
```

```python
len(count_of_words)
```

```python
sorted_words=sorted(list(count_of_words.values()),reverse=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.loglog(sorted_words)
plt.xlabel('rank of words')
plt.ylabel('frequency')
plt.show()
```

```python
plt.hist(sorted_words,bins=50)
plt.show()
```

```python
plt.hist(sorted_words,bins=50,log=True)
plt.show()
```

# lets do sentiment analysis

```python
lposr=[(after_removal_of_stopwords(movie_reviews.words(fileids=[f])),'pos') for f in posr]
```

```python
print(lposr[2])
```

```python
lneg=[(after_removal_of_stopwords(movie_reviews.words(fileids=[f])),'neg') for f in negr]
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
split=800
clf=NaiveBayesClassifier.train(lposr[:split]+lneg[:split])
```

```python
nltk.classify.accuracy(clf,lposr[split:]+lneg[split:])
```

```python
nltk.classify.accuracy(clf,lposr[:split]+lneg[:split])
```
